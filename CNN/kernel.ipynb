```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import io, color
```

```python
path = './files'
```

```python
img = io.imread(f"{path}/yo.png")
img_gray = color.rgb2gray(img)
```

```python
plt.imshow(img_gray)
```

```python
kernel = np.array([[-1,0,1], [-1,0,1], [-1,0,1]])
kernel2 = np.array([[-1,-1,-1], [0,0,0], [1,1,1]])
kernel3 = np.array([[1,1,1], [1,1,1], [1,1,1]])
kernel4 = np.array([[-1,-1,-1], [-1,-1,-1], [-1,-1,-1]])
```

```python
img_conv = nd.convolve(img_gray, kernel4)
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,10))
axes[0].imshow(img_gray, cmap='gray')
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(img_conv, cmap='gray')
axes[1].set_title('Convolucion')
axes[1].axis('off')
```
