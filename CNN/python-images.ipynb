```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
path = './files'
```

```python
img = io.imread(f'{path}/Mara.jpeg')
```

```python
plt.imshow(img)
```

```python
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
```

```python
aux_dim = np.zeros([800, 600])
```

```python
img_r = np.dstack((r, aux_dim, aux_dim)).astype(np.uint8)
img_g = np.dstack((aux_dim, g, aux_dim)).astype(np.uint8)
img_b = np.dstack((aux_dim, aux_dim, b)).astype(np.uint8)
```

```python
img_m_g = img_g[400:, :300]
img_m_r = img_r[:400, 300:]
img_m_b = img_b[400:, 300:]
img_m = img[:400, :300]
total_img_1 = np.vstack((img_m, img_m_g,))
total_img_2 = np.vstack((img_m_r, img_m_b))
total_img = np.concatenate((total_img_1, total_img_2), axis=1)
```

```python
plt.imshow(total_img)
```
