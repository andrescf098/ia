```python
v1 = np.array([0,1])
v2 = np.array([1,0])

graficarVectores([v1, v2], ['blue', 'red'])

plt.xlim(-0.25, 2)
plt.ylim(-0.25, 2)
```

```python
A = np.array([[2, 0], [0, 2]])
v1_trans = A.dot(v1)
v2_trans = A.dot(v2)

graficarVectores([v1_trans, v2_trans], ['orange', 'green'])

plt.xlim(-0.25, 3)
plt.ylim(-0.25, 3)
```

```python
det_A = np.linalg.det(A)
print(det_A)
```

```python
area_transform = abs((v1_trans[0] - v2_trans[0])) * abs((v1_trans[1] - v2_trans[1]))
print(area_transform)
```

```python
B = A * [-1, 1]
print(np.linalg.det(B))
```
