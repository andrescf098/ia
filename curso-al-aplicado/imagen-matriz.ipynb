```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')
imagen = Image.open("imagenes/gato.jpg")

plt.imshow(imagen)
```

```python
image_gr = imagen.convert("LA")
plt.imshow(image_gr)
```

```python
imagen_mat = np.array(list(image_gr.getdata(band=0)), float)
print(imagen_mat)
```

```python
imagen_mat.shape = (image_gr.size[1], image_gr.size[0])
print(imagen_mat)
```

```python
print(imagen_mat.shape)
```

```python
plt.imshow(imagen_mat, cmap='gray')
```

```python
imagen_mat_2 = imagen_mat / 10
plt.imshow(imagen_mat_2)
plt.show()
```

```python
print(np.max(imagen_mat_2))
print(np.max(imagen_mat))
print(np.min(imagen_mat_2))
print(np.min(imagen_mat))
```
