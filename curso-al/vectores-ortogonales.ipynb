# **Vectores ortogonales**

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([0,0,2,2])
y = np.array([0,0,2,-2])

plt.quiver([x[0], y[0]], [x[1], y[1]], [x[2], y[2]], [x[3], y[3]], angles='xy', scale_units='xy', scale=1, color=sns.color_palette())

plt.xlim(-2, 4)
plt.ylim(-3, 3)

plt.show()
```

```python
v1 = np.array([[2,2]])
v2 = np.array([[2,-2]])

print(v1.dot(v2.T))
```

```python
print(np.linalg.norm(v1))
print(np.linalg.norm(v2))
```

```python
v1 = np.array([[1,0]])
v2 = np.array([[0,-1]])

print(v1.dot(v2.T))
```

```python
print(np.linalg.norm(v1))
print(np.linalg.norm(v2))
```
