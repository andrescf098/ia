```python
import pandas as pd
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=iris, hue= 'species')
```

```python
sns.jointplot(x='sepal_width', y='petal_width', data=iris, kind='reg')
```

```python
sns.boxplot(x='species', y='sepal_length', data=iris, hue= 'species')
```

```python
sns.barplot(x='species', y='sepal_length', data=iris, hue= 'species')
```
