```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

month = np.array(["E", "F", "M", "A", "Ma", "J", "Ju", "Ag", "S", "O", "N", "D"])
sales = np.array([100, 120, 140, 160, 180, 200, 220, 240, 220, 280, 300, 320])

plt.figure(figsize=(8,6))
plt.plot(month, sales, marker="o", color="red")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))

hours = np.array([i for i in range(0, 100)])
exam_student_1 = np.array([np.random.randint(20, 100) for i in range(100)])
exam_student_2 = np.array([np.random.randint(20, 100) for i in range(100)])
plt.scatter(hours, exam_student_1, marker='o', color="blue", linestyle='-', label="Student 1")
plt.scatter(hours, exam_student_2, marker='s', color="red", linestyle='-', label="Student 2")
plt.title("Exam Score vs Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))

hours = np.array([i for i in range(0, 100)])
exam_student_1 = np.array([np.random.randint(2, 10)*hours[i]/10 for i in range(100)])
exam_student_2 = np.array([np.random.randint(2, 10)*hours[i]/10 for i in range(100)])
plt.plot(hours, exam_student_1, marker='o', color="blue", linestyle='-', label="Student 1", linewidth=2)
plt.plot(hours, exam_student_2, marker='s', color="red", linestyle='-', label="Student 2" , linewidth=2)
plt.title("Exam Score vs Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
categories = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
values = [np.random.randint(100, 1000) for i in range(5)]
plt.bar(categories, values, color="skyblue", label="Sales")
plt.title("Product Sales")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
categories = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
values = np.random.rand(5)
values = [round(i/sum(values)*100, 2) for i in values]
values[-1] = 100 - sum(values[:-1])
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("tab10"))
plt.axis('equal')
plt.show()
```
